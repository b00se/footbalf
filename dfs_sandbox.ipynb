```python
import nfl_data_py as nfl
import pandas as pd
```

```python
df_train = nfl.import_weekly_data([2019, 2020, 2021, 2022])
df_test = nfl.import_weekly_data([2023])
```

```python
nfl.see_weekly_cols()
```

```python
pd.set_option('display.max_columns', None)
df_train.describe(include="all")
```

```python
# find missing values clusters by time period
df_train.groupby(['season', 'week']).apply(lambda x: x.isnull().sum())
```

```python
#analyze missing values by position
df_train.groupby('position').apply(lambda x: x.isnull().sum())
```

```python
# drop any rows with irrelevant positions
df_train = df_train[df_train['position'].isin(['QB', 'RB', 'WR', 'TE', 'K'])]
df_train.groupby('position').apply(lambda x: x.isnull().sum())
```

```python
# Columns to fill with zeros
columns_to_fill = ['passing_epa', 'pacr', 'dakota', 'rushing_epa', 'receiving_epa', 
                   'racr', 'target_share', 'air_yards_share', 'wopr']

# Filling missing values in these columns with zeros
df_train.loc[:, columns_to_fill] = df_train[columns_to_fill].fillna(0)
df_train[columns_to_fill].isnull().sum()
```

```python
# filter out playoff games and last week of season
df_train = df_train[df_train['season_type'] == 'REG']
df_train = df_train[~((df_train['season'] < 2021) & (df_train['week'] == 17)) & ~((df_train['season'] >= 2021) & (df_train['week'] == 18))]
df_train.groupby('season')['week'].max()
```

```python
# get schedules
df_schedules = nfl.import_schedules([2019, 2020, 2021, 2022])
df_schedules.head(1)
```

```python
# merge df_train with home and away games
merged_home = pd.merge(df_train, df_schedules, left_on=['season', 'week', 'recent_team'], right_on=['season', 'week', 'home_team'], how='left')
merged_away = pd.merge(df_train, df_schedules, left_on=['season', 'week', 'recent_team'], right_on=['season', 'week', 'away_team'], how='left')

# drop records where game_id is NaN
merged_home = merged_home[merged_home['game_id'].notna()]
merged_away = merged_away[merged_away['game_id'].notna()]

# concatenate results and drop duplicate columns
merged_df = pd.concat([merged_home, merged_away]).reset_index(drop=True)

```

```python
# Create a unique identifier in both datasets
df_train['unique_key'] = df_train['player_id'] + '_' + df_train['season'].astype(str) + '_' + df_train['week'].astype(str)
merged_df['unique_key'] = merged_df['player_id'] + '_' + merged_df['season'].astype(str) + '_' + merged_df['week'].astype(str)

# Find unique keys present in df_train but not in merged_df
missing_keys = set(df_train['unique_key']) - set(merged_df['unique_key'])

# Extract the missing records
missing_records = df_train[df_train['unique_key'].isin(missing_keys)]

# Display the missing records
missing_records.describe(include="all")
```

```python
df_merged_players = merged_df.copy()

# get all player records for 2019 where recent_team is LV and change it to OAK
lv_players = df_train[(df_train['season'] == 2019) & (df_train['recent_team'] == 'LV')]
lv_players.loc[:, 'recent_team'] = 'OAK'

# merge schedule data for 2019
merged_home = pd.merge(lv_players, df_schedules, left_on=['season', 'week', 'recent_team'], right_on=['season', 'week', 'home_team'], how='left')
merged_away = pd.merge(lv_players, df_schedules, left_on=['season', 'week', 'recent_team'], right_on=['season', 'week', 'away_team'], how='left')
merged_home = merged_home[merged_home['game_id'].notna()]
merged_away = merged_away[merged_away['game_id'].notna()]

df_merged = pd.concat([merged_home, merged_away]).reset_index(drop=True)
# combine with df_merged_players
df_merged_players = pd.concat([df_merged_players, df_merged]).reset_index(drop=True)
df_merged_players[25:35]
```
